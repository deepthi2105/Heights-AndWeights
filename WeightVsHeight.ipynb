```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#read the csv file
dataset=pd.read_csv("weight-height.csv")
```

```python
#shape of the dataset
dataset.shape
```

```python
#displaying the first 5 rows
dataset.head()
```

```python
#describe the dataset(mean,count,standard deviation,etc)
dataset.describe()
```

```python
#check if any null values are present
dataset.isnull().sum()
```

```python
#analysing data via boxplot
import seaborn as sns
sns.boxplot(data=dataset,width=0.5,fliersize=5)
sns.set(rc={'figure.figsize':(10,5)})
```

```python
#analysing data via histogram
dataset.hist()
```

```python
plt.scatter(dataset.Height,dataset.Weight)
```

```python
#Assigning female as 1 and male as 0
dataset['Gender'].replace('Female',0,inplace=True)
dataset['Gender'].replace('Male',1,inplace=True)
```

```python
dataset.head()
```

```python
#rounding the data to 3 decimal places
dataset.round({"Height":3,"Weight":3})
```

```python
X=dataset.iloc[:,0:2]
Y=dataset.iloc[:,2]
print(X.shape)
print(Y.shape)
```

```python
#Splitting the dataset into train and test sets
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#Applying Linear Regression
clf=LinearRegression()
clf.fit(X_train,Y_train)
pred=clf.predict(X_test)
```

```python
#accuracy
clf.score(X_test,Y_test)
```
